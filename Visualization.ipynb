```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[50,51,52,78,42,20,21]
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(x,y, color="green")
```

```python
days=[1,2,3,4]
max_t=[50,51,52,48]
min_t=[7,12,15,20]
avg=35,30,45,33
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(days,max_t, label="max")
plt.plot(days,min_t, label="min")
plt.plot(days,avg, label="avg")
plt.legend(loc="best")
plt.grid()
```
